**Question 3**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a sample dataset
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C'],
        'Value': [10, 12, 8, 9, 15, 14, 11, 10, 16]}
df = pd.DataFrame(data)

# Create a box plot to compare categories based on their values
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Comparison of Categories based on Values')
plt.show()

# Create a violin plot to compare categories based on their values
sns.violinplot(x='Category', y='Value', data=df)
plt.title('Comparison of Categories based on Values (Violin Plot)')
plt.show()

# Create a bar plot to compare the mean values of different categories
sns.barplot(x='Category', y='Value', data=df, estimator=np.mean)
plt.title('Mean Value Comparison of Categories')
plt.show()

# Create a swarmplot to show individual data points and compare categories
sns.swarmplot(x='Category', y='Value', data=df)
plt.title('Comparison of Categories with Individual Data Points')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a sample dataset
np.random.seed(42)  # Set a seed for reproducibility
data = {'Category': ['A'] * 50 + ['B'] * 50 + ['C'] * 50,
        'Value': np.concatenate([np.random.normal(10, 2, 50),
                                 np.random.normal(15, 3, 50),
                                 np.random.normal(20, 4, 50)])}
df = pd.DataFrame(data)

# Visualize the distribution of 'Value' across different 'Category' using a boxplot
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Value across Categories (Boxplot)')
plt.show()

# Visualize the distribution of 'Value' across different 'Category' using a violin plot
sns.violinplot(x='Category', y='Value', data=df)
plt.title('Distribution of Value across Categories (Violin Plot)')
plt.show()

# Visualize the distribution of 'Value' across different 'Category' using a swarmplot
sns.swarmplot(x='Category', y='Value', data=df)
plt.title('Distribution of Value across Categories (Swarmplot)')
plt.show()
```

**Question 5**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a synthetic dataset with correlated features
np.random.seed(42)
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 + np.random.randn(n_samples) * 0.1  # Introduce some noise
x3 = x1 * 2 + np.random.randn(n_samples) * 0.2  # Higher correlation with x1
data = pd.DataFrame({'X1': x1, 'X2': x2, 'X3': x3})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Pivot the DataFrame to create a matrix where rows are months, columns are days, and values are sales
sales_matrix = df.pivot_table(index='Month', columns='Day', values='Sales')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(sales_matrix, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales Variation by Month and Day")
plt.xlabel("Day")
plt.ylabel("Month")
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='Viridis')])
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable notebook output
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

output_notebook()

# Generate random data
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.randint(5, 20, n)
colors = np.random.choice(['red', 'green', 'blue', 'orange'], n)

# Create a DataFrame for easier handling of data
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot", x_axis_label="x", y_axis_label="y")

# Add a scatter glyph
p.scatter(x='x', y='y', source=data, size='sizes', color='colors', legend_field='colors')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Generate the histogram data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Bokeh Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add a quad glyph for the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```
